### Bloco A - Operadores e variáveis
- Dados ano de nascimento e morte de 4 autores (Machado (1839-1908), Eça (1845-1900), Guimarães Rosa (1908-1967), Clarice (1920-1977)), calcule a idade de cada um em f-string.
- Machado publicou um livro em 1881 e outro em 1899.
Entre o primeiro e o último livro, há 2 intervalos.
Calcule em variáveis:

    1. Quantos anos se passaram de 1881 até 1899.

    2. Quanto vale cada intervalo, se esse tempo for dividido em 2 partes.

- Se um romance tem 400 páginas e o aluno lê 15 por dia, quantos dias inteiros precisa? E quantas páginas sobram no último dia?

```python
# Anos de nascimento e morte
M_nas = 1839; M_mor = 1908
E_nas = 1845; E_mor = 1900
G_nas = 1908; G_mor = 1967
C_nas = 1920; C_mor = 1977
# Idade de cada um
print(f"Machado morreu com {M_mor - M_nas} anos.")
print(f"Eça morreu com {E_mor - E_nas} anos.")
print(f"Guimarães morreu com {G_mor - G_nas} anos.")
print(f"Clarice morreu com {C_mor - C_nas} anos.")
```

```python
# Intervalo entre publicações
prim_pub = 1881
ult_pub = 1899
#intervalo entre primeira e última publicação
intervalo_pub = ult_pub - prim_pub
# intervalo em duas partes
cada_intervalo = intervalo_pub/2
print(f"Entre o primeiro e o último livro passaram {ult_pub - prim_pub} anos.")
print(f"Dividido em duas partes, cada intervalo de publicação é de {cada_intervalo} anos.")
```

```python
# Páginas por dia + sobra no último dia
total_pag = 400
pag_dia = 15
# Quantos dias inteiros para ler?
dias_inteiros = total_pag//pag_dia
# Páginas que sobraram no último dia
sobra = total_pag % pag_dia
print(f"O aluno precisou de {dias_inteiros} dias inteiros para ler um livro de 400 páginas, lendo 15 páginas por dia.")
print(f"No último dia, sobraram {sobra} páginas para o aluno ler.")
```
